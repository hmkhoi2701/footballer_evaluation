## Now begins the table handling

```python
transfermarkt_url = [competitions_read.loc[competitions_read.domestic_league_code == x].iloc[0,7] 
       for x in ["GB1", "IT1", "ES1", "FR1", "L1"]]
transfermarkt_url
```

```python
PATH = Service("C:\Program Files (x86)\chromedriver.exe")

driver = webdriver.Chrome(service = PATH)

team_names = []

# extract only teams currently in the league
for league_url in transfermarkt_url:
    driver.get(league_url)

    
    # get number of teams in the league
    league_details = driver.find_element("xpath", "(//tbody)").find_element("xpath", "(//tr)[2]").get_attribute('innerHTML')
    league_details.replace(' ','')
    start_stamp = league_details.find('<td>')+5
    end_stamp = league_details.find('teams',start_stamp)
    no_of_teams = int(league_details[start_stamp:end_stamp].strip())

    
    # get name of each team, matching club table convention
    search = driver.find_element("xpath", "(//tbody)[4]")
    for _ in range (1, no_of_teams+1):
        detail = search.find_element("xpath","(//td[@class='hauptlink no-border-links'])[%i]" %_).get_attribute('innerHTML')
        start_stamp = detail.find('/')+1
        end_stamp = detail.find('/',start_stamp)
        team_names.append(detail[start_stamp:end_stamp])

driver.close()

len(team_names)
```

```python
filter_table = club_read.loc[(club_read.domestic_competition_id.isin(["GB1", "IT1", "ES1", "FR1", "L1"])) & (club_read.name.isin(team_names))].iloc[:,[0,1,2,3,-1]]
```

```python
club_table = pd.DataFrame({'name':team_names}).merge(filter_table)
club_table
```

```python
for i in range(len(club_table.club_id)):
    if i == 0:
        mc_players = players_read.loc[(players_read.current_club_id == club_table.club_id[i])&(players_read.last_season == 2021)]
    else:
        club_players = players_read.loc[(players_read.current_club_id == club_table.club_id[i])&(players_read.last_season == 2021)]
        mc_players = pd.concat([mc_players,club_players])
        
mc_players = mc_players.iloc[:,[0,2,3,4,5,7,8,9,10]]
```

```python
mc_players
```

```python
player_past = pd.merge(mc_players, players_valuations_read, on = ['player_id'])
player_past
```

```python
def endseason(date):
    return abs(sum(int(date[5:7])-i for i in [5,6,7,8,9]))

def date_extract(datelist):
    yearlist = {}
    for date in datelist:
        if not date[:4] in yearlist:
            yearlist[date[:4]] = date
        elif endseason(date) < endseason(yearlist[date[:4]]):
            yearlist[date[:4]] = date
    return yearlist
```

```python
arr = pd.unique(player_past.name)
for i in range(len(arr)):
    year_player = player_past.loc[player_past.name == arr[i]]
    if i == 0:
        fin = year_player.loc[year_player.date.isin(date_extract(year_player.date).values())]
    else:
        fin = pd.concat([fin, year_player.loc[year_player.date.isin(date_extract(year_player.date).values())]])
fin
```

```python
fin["date"] = fin["date"].apply(lambda d: d[:4]).astype('int64')
```

```python
fin = fin.loc[fin.date != 2023]
fin
```

```python
player_performance = pd.read_csv("player_performance.csv")
player_performance.head()
```

```python
data_table = pd.merge(fin,player_performance,on = ['name','date'])
data_table = pd.merge(data_table,club_table,left_on = ['current_club_id'],right_on=['club_id'])
data_table = data_table.iloc[:,list(range(2,19))+[-3,-2]]
data_table
```

```python
df = data_table.columns.tolist()
df = [df[-1],df[-2]]+df[:-2]
data_table = data_table[df]
data_table["domestic_competition_id"] = data_table["domestic_competition_id"].apply(lambda id: "EPL" if id=='GB1' else ("Serie A" if id=='IT1' else("La Liga" if id=='ES1' else("League 1" if id=='FR1' else "Bundesliga"))))
```

```python
data_table.rename(columns = {"pretty_name_y":"club","pretty_name_x":"name","domestic_competition_id":"league"},inplace = True)
data_table
```

```python
def age_at(y):
    return y['date'] - int(y['date_of_birth'][-4:])
data_table['age'] = data_table.apply(age_at,axis = 1)
data_table
```

```python
data_table.to_csv("data_table.csv",index = False)
```
