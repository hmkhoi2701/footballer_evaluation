```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
PATH = Service("C:\Program Files (x86)\chromedriver.exe")

driver = webdriver.Chrome(service = PATH)
```

```python
club_read = pd.read_csv("raw_datasets\clubs.csv")
club_read.head()
```

```python
competitions_read = pd.read_csv("raw_datasets\competitions.csv")
competitions_read.head()
```

```python
players_read = pd.read_csv("raw_datasets\players.csv")
players_read.head()
```

```python
players_valuations_read = pd.read_csv("raw_datasets\player_valuations.csv")
players_valuations_read.head()
```

```python
url = [competitions_read.loc[competitions_read.domestic_league_code == x].iloc[0,7] 
       for x in ["GB1", "IT1", "ES1", "FR1", "L1"]]
url
```

```python
PATH = Service("C:\Program Files (x86)\chromedriver.exe")

driver = webdriver.Chrome(service = PATH)

team_names = []

# extract only teams currently in the league
for league_url in url:
    driver.get(league_url)

    
    # get number of teams in the league
    league_details = driver.find_element("xpath", "(//tbody)").find_element("xpath", "(//tr)[2]").get_attribute('innerHTML')
    league_details.replace(' ','')
    start_stamp = league_details.find('<td>')+5
    end_stamp = league_details.find('teams',start_stamp)
    no_of_teams = int(league_details[start_stamp:end_stamp].strip())

    
    # get name of each team, matching club table convention
    search = driver.find_element("xpath", "(//tbody)[4]")
    for _ in range (1, no_of_teams+1):
        detail = search.find_element("xpath","(//td[@class='hauptlink no-border-links'])[%i]" %_).get_attribute('innerHTML')
        start_stamp = detail.find('/')+1
        end_stamp = detail.find('/',start_stamp)
        team_names.append(detail[start_stamp:end_stamp])

driver.close()

team_names
```

```python
filter_table = club_read.loc[(club_read.domestic_competition_id == 'GB1') & (club_read.name.isin(team_names))].iloc[:,[0,1,2,3,-1]]
```

```python
epl_table = pd.DataFrame({'name':team_names}).merge(filter_table)
epl_table
```

```python
mc_players = players_read.loc[(players_read.current_club_id == epl_table.iloc[0].club_id)&(players_read.last_season == 2021)]
mc_players
```

```python
player_past = pd.merge(mc_players, players_valuations_read, on = ['player_id'])
player_past.loc[(player_past.player_id == mc_players.iloc[0].player_id)]
```
