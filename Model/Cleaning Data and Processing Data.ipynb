# 1.Import library and read data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data_table.csv")
data.head()
```

```python
# remove redudance columns
data = data.drop(columns = ['name','country_of_birth','date_of_birth','name_x','club','sub_position'])
data.head()
```

# 3 .Handling missing values

```python
# check null values in data
data.isnull().sum()
```

### we have foot feature that has missing values , so we use mode method for foot feature

```python
index = 0;
for i in data['foot'].isnull():
    if i :
        data['foot'][index] = data['foot'].mode()[0]
    index = index + 1
```

# 3. Convert category data into numerical data

```python
# check data types
data.dtypes
```

```python
# foot attribute
for i in range(len(data['foot'])):
    if data['foot'][i] =='Right':
        data['foot'][i] = 1
    elif data['foot'][i] == 'Left':
        data['foot'][i] = 2
    else:
        data['foot'][i] = 3
data["foot"] = data["foot"].astype(str).astype(float)

# league attribute
for i in range(len(data['league'])):
    if data['league'][i] == 'EPL':
        data['league'][i] = 5
    elif data['league'][i] == 'La Liga':
        data['league'][i] = 4
    elif data['league'][i] == 'Bundesliga':
        data['league'][i] = 3
    elif data['league'][i] == 'Serie A':
        data['league'][i] = 2
    else:
        data['league'][i] = 1
data["league"] = data["league"].astype(str).astype(float)

```
